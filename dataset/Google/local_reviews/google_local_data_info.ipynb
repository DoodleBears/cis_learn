```python
import pandas as pd
import gzip
import json
import ast
```

```python
def getDF(path):
    with gzip.open(path, 'rb') as g:
        i = 0
        df = {}
        for line in g:
            # print(l)
            jsonLine = line.decode()
            # print(jsonFile)
            
            json_temp = ast.literal_eval(jsonLine)
            df[i] = json_temp
            i += 1
            if i > 1000000 : break
        return pd.DataFrame.from_dict(df, orient='index')
```

```python
# dataset: https://cseweb.ucsd.edu//~jmcauley/datasets.html
place_df = getDF('places.clean.json.gz')
user_df = getDF('users.clean.json.gz')
review_df = getDF('reviews.clean.json.gz')
```

```python
review_place_df = pd.merge(left=review_df, right=place_df)
review_place_df
```

```python
user_place_review_df = pd.merge(left=review_place_df, right=user_df)
user_place_review_df = user_place_review_df.drop(['userName', 'reviewText', 'previousPlaces'], axis=1)
rename_map = {'name': 'placeName', 'phone': 'placePhone', 'hours':'openHours'}
user_place_review_df = user_place_review_df.rename(columns=rename_map)
user_place_review_df = user_place_review_df.sort_values('reviewerName')
user_place_review_df = user_place_review_df.reset_index()
user_place_review_df = user_place_review_df.drop(['index'], axis=1)
user_place_review_df
```
