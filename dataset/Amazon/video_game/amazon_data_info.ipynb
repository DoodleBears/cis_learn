```python
import pandas as pd
import gzip
import json
```

```python
def parse(path):
  g = gzip.open(path, 'rb')
  for l in g:
    yield json.loads(l)

def getDF(path):
  i = 0
  df = {}
  for d in parse(path):
    df[i] = d
    i += 1
  return pd.DataFrame.from_dict(df, orient='index')
# dataset: https://cseweb.ucsd.edu//~jmcauley/datasets.html
df = getDF('Video_Games_5.json.gz')
df
```

```python
df = df.drop(labels=['style', 'vote', 'image'], axis=1)
df
# asin: product id
```
