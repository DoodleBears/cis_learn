```python
print(len(merge_train_df.columns))
X_full = merge_train_df.drop(columns='target', axis=1)
y_full = pd.DataFrame(merge_train_df.target)
```

```python
X_full = X_full.drop(columns=['genre_ids'])
# X_full = X_full.drop(columns=['song_length', 'composer', 'lyricist', 'registration_init_time'])
# X_full = X_full.drop(columns=['expiration_date', 'bd', 'registered_via'])
X_full.head(3)
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X_full, y_full, train_size=0.001, random_state=0)
```

```python
X_train.to_csv('X_train.csv', header=None,index=None)
X_valid.to_csv('X_valid.csv', header=None,index=None)
y_train.to_csv('y_train.csv', header=None,index=None)
y_valid.to_csv('y_valid.csv', header=None,index=None)
```

```python
merge_train, merge_test = train_test_split(merge_train_df, train_size=0.8, test_size=0.2, random_state=0)

merge_train.to_csv('merge_train.csv', header=None,index=None)
merge_test.to_csv('merge_test.csv', header=None,index=None)
```

```python
import xlearn as xl
```

```python
xdm_train = xl.DMatrix(X_train, y_train)
xdm_test = xl.DMatrix(X_valid, y_valid)
```

```python
# 开始 train
# Training task
fm_model = xl.create_fm()  # Use factorization machine
# fm_model.setNoBin()
fm_model.setTrain(xdm_train)    # Training data
fm_model.setValidate(xdm_test)  # Validation data
# param
param = {'task':'binary', 'lr':0.2, 
         'lambda':0.002, 'metric':'acc'}

# Start to train
# The trained model will be stored in model.out
# fm_model.fit(param, './fm_predict_model_dm.out')
```

```python
# Prediction task
fm_model.setTest(xdm_test)  # Test data
fm_model.setSigmoid()  # Convert output to 0-1

# Start to predict
# res = fm_model.predict("./fm_predict_model_dm.out", "./fm_predict_output.txt")
```
