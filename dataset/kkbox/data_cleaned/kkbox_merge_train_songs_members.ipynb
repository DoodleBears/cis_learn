```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import json
```

```python
members = pd.read_csv('members.csv')
songs = pd.read_csv('songs.csv')
train = pd.read_csv('train.csv')
```

```python
members_df = members
songs_df = songs
train_df = train
```

```python
X_full = pd.merge(train_df, songs_df)
X_full
```

```python
X_full = pd.merge(X_full, members_df)
X_full
```

```python
X_full.to_csv('merge_train_songs_members.csv', index=None)
```

```python
# 将song_id 转化为 int
song_id_list = np.array(X_full.song_id.unique())
song_id_map = {}
song_id_dict = {}
for i in range(len(song_id_list)):
    temp_str = str(song_id_list[i])
    song_id_map.update({temp_str:i})
    song_id_dict.update({i:temp_str})

X_full['song_id'] = X_full.song_id.map(song_id_map)
song_id_map
```

```python
f = open("../dict/song_id_dict.json", "w")
json.dump(song_id_dict, f)
f.close()
```

```python
X_full
```

```python
# 将msno 转化为 int
msno_list = np.array(X_full.msno.unique())
msno_map = {}
msno_dict = {}
for i in range(len(msno_list)):
    temp_str = str(msno_list[i])
    msno_map.update({temp_str:i})
    msno_dict.update({i:temp_str})

X_full['msno'] = X_full.msno.map(msno_map)
msno_map
```

```python
f = open("../dict/msno_dict.json", "w")
json.dump(msno_dict, f)
f.close()
```
