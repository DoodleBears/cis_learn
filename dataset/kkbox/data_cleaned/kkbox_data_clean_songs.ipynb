```python
import pandas as pd
import numpy as np
```

```python
def showDFBasicInfo(df):
    # check if duplicated
    print('is duplicated exist: ' + str(df[df.duplicated()]))
    # check if null
    np.where(pd.isnull(df))
    # print type of data in DataFrame 
    print(df.dtypes)
```

```python
songs = pd.read_csv('../data_origin/songs.csv')
```

```python
songs_df = songs
songs_df.head(3)
songs_df.info()
```

```python
songs_df.head()
```

```python
# map gender to binary
songs_df = songs_df.drop(index=songs_df[(songs_df.genre_ids.isnull()) | (songs_df.composer.isnull()) | (songs_df.lyricist.isnull()) | (songs_df.song_length.isnull()) | (songs_df.song_id.isnull()) | (songs_df.language.isnull())].index)
# reset index
songs_df = songs_df.reset_index()
songs_df = songs_df.drop(['index'],axis=1)
print(len(songs_df.artist_name.unique()))
songs_df.artist_name.unique()
```

```python
songs_df.head(3)
```

```python
artist_name_list = np.array(songs_df.artist_name.unique())
artist_name_map = {}
for i in range(len(artist_name_list)):
    temp_str = str(artist_name_list[i])
    artist_name_map.update({temp_str:i})

songs_df['artist_name'] = songs_df.artist_name.map(artist_name_map)
artist_name_map
```

```python
songs_df.head(3)
```

```python
composer_list = np.array(songs_df.composer.unique())
composer_map = {}
for i in range(len(composer_list)):
    temp_str = str(composer_list[i])
    composer_map.update({temp_str:i})

songs_df['composer'] = songs_df.composer.map(composer_map)
composer_map
```

```python
songs_df.head(3)
```

```python
lyricist_list = np.array(songs_df.lyricist.unique())
lyricist_map = {}
for i in range(len(lyricist_list)):
    temp_str = str(lyricist_list[i])
    lyricist_map.update({temp_str:i})

songs_df['lyricist'] = songs_df.lyricist.map(lyricist_map)
lyricist_map
```

```python
songs_df.head(3)
```

```python
songs_df['language'] = songs_df.language.astype(int)
songs_df.language.unique()
```

```python
songs_df.head(3)
```

```python
songs_df.genre_ids.unique()
```

```python
new_genre_ids_list = []
for i in range(len(songs_df)):
    new_genres = []
    # extract every anime's genres
    str_genres = str(songs_df.genre_ids[i]) 
    arr_splitted = str_genres.split('|')
    # print(arr_splitted[0])
    new_genres.append(songs_df.song_id[i])
    new_genres.append(arr_splitted[0])
    new_genre_ids_list.append(new_genres)
```

```python

new_genres_ids_solo_array = np.array(new_genre_ids_list)
genres_ids_solo_columns = ['song_id', 'genre_ids_solo']
genres_ids_solo_df = pd.DataFrame(data=new_genres_ids_solo_array, columns=genres_ids_solo_columns)
```

```python
genres_ids_solo_df
```

```python
songs_df
```

```python
songs_df = pd.merge(songs_df, genres_ids_solo_df)
songs_df
```

```python
songs_df.to_csv('songs.csv', index=None)
```
