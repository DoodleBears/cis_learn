```python
import pandas as pd
import numpy as np
import sys
```

```python
merge_train_100 = pd.DataFrame()
merge_train_100= pd.read_csv('fm_train_data_100.txt', header=None ,sep=' ')
merge_train_100
```

```python
def process_line(line, train_full_len):
    sys.stdout.write('\r')
    process_str = '['
    for i in range(line):
        process_str = process_str + '|'
    for i in range(line, 101):
        process_str = process_str + ' '

    process_str = process_str + ']{}/{}'.format(line,train_full_len)
    sys.stdout.write(process_str)
```

```python
merge_train_100_size = len(merge_train_100)
for rate in range(1, 10):
    merge_train_split = pd.DataFrame()
    split_size = merge_train_100_size // 100 * rate
    # print(split_size)
    process_line(rate, 100)
    merge_train_split = merge_train_100[:split_size]
    merge_train_split.to_csv('fm_train_data_{}.txt'.format(rate), header=None, index=None, sep=' ')
```

```python
for rate in range(10, 100, 10):
    merge_train_split = pd.DataFrame()
    split_size = merge_train_100_size // 100 * rate
    # print(split_size)
    process_line(rate, 100)
    merge_train_split = merge_train_100[:split_size]
    merge_train_split.to_csv('fm_train_data_{}.txt'.format(rate), header=None, index=None, sep=' ')
```
