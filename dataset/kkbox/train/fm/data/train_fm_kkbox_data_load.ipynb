```python
import sys
def process_line(line, train_full_len):
    sys.stdout.write('\r')
    process_str = '['
    for i in range(line):
        process_str = process_str + '|'
    for i in range(line, 101):
        process_str = process_str + ' '

    process_str = process_str + ']{}/{}'.format(index,train_full_len)
    sys.stdout.write(process_str)
```

```python
f = open("fm_test_data_{}.txt".format(int(test_size*100)), "wb")

test_full_len = len(test_full)
one_percent_count = test_full_len // 100

str_temp = ''
index_temp = 0
line = 0
for index in range(test_full_len):
    row = test_full.iloc[index]
    str_temp = str(row[0])
    for col in range(1, len(test_full.columns)):
        str_temp = str_temp + ' {}:1'.format(row[col])
    str_temp = str_temp + '\n'
    f.write(str_temp.encode('utf-8'))
    index_temp += 1
    if index_temp >= one_percent_count:
        line += 1
        index_temp = 0
        process_line(line, test_full_len)

f.close()
```
