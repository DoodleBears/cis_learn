```python
merge_train_df = merge_train_df.drop(['song_length'], axis=1)
merge_train_df = merge_train_df.drop(['genre_ids'], axis=1)
merge_train_df = merge_train_df.drop(['registration_init_time', 'expiration_date'], axis=1)
merge_train_df = merge_train_df.drop(['composer', 'lyricist'], axis=1)
# merge_train_df = merge_train_df.drop(['artist_name'], axis=1)
# merge_train_df = merge_train_df.drop(['genre_ids_solo'], axis=1)
# merge_train_df = merge_train_df.drop(['registration_year'], axis=1)
# merge_train_df = merge_train_df.drop(['song_id', 'msno'], axis=1)
# merge_train_df = merge_train_df.drop(['language'], axis=1)
# merge_train_df = merge_train_df.drop(['bd'], axis=1)
# merge_train_df = merge_train_df.drop(['city', 'registered_via'], axis=1)
# merge_train_df = merge_train_df.drop(['source_system_tab', 'source_screen_name'], axis=1)
# merge_train_df = merge_train_df.drop(['source_type'], axis=1)
# merge_train_df = merge_train_df.drop(['registration_year'], axis=1)
```

```python
# TODO 将feature:value 对应的 value 改为 feature 在 features 中的比例
feature_s_value_map = {}
for index in range(len(merge_train_df.columns)):
    # value_map = {}
    value_map = pd.DataFrame(merge_train_df[[merge_train_df.columns[index]]].value_counts())
    field_name = str(merge_train_df.columns[index])
    value_map = value_map.reset_index()
    
    rename_map = {0: 'value'}
    value_map = value_map.rename(columns = rename_map)
    print(field_name)
    # print(value_map)
    value_count = merge_train_df[[merge_train_df.columns[index]]].count()
    feature_value_map = {}
    for i in range(len(value_map)):
        # print(end='')
        # print(str(merge_train_df.columns[index]) +': ' + str(value_map[i]))
        feature_value_map.update({value_map[field_name][i]: value_map.value[i] / value_count})
    # print(feature_value_map)
    feature_s_value_map.update({index: feature_value_map})
# feature_s_value_map
```

```python
train_size = 0.01
test_size = 0.2
train_full, test_full = train_test_split(merge_train_df, train_size=train_size, test_size=test_size, random_state=0)
```

```python
f = open("ffm_train_data_{}.txt".format(int(train_size*100)), "wb")
train_full_list = []
str_temp = ''
train_full_len = len(train_full)
train_1_len = train_full_len // 100
index_temp = 0
line = 0
for index in range(train_full_len):
    row = train_full.iloc[index]
    str_temp = str(row[0])
    
    for col in range(1, len(train_full.columns)):
        # feature = feature_s_value_map.get(col)
        # print(feature[row[col]][0])
        # str_temp = str_temp + ' {}:{}:{}'.format(col, row[col], np.round(feature[row[col]][0], 5))
        str_temp = str_temp + ' {}:{}:1'.format(col, row[col])
    str_temp = str_temp + '\n'
    # train_full_list.append(str_temp.encode('ascii'))
    # f.write(str_temp)
    f.write(str_temp.encode('utf-8'))

    index_temp += 1
    if index_temp >= train_1_len:
        sys.stdout.write('\r')
        line += 1
        index_temp = 0
        process_str = '['
        for i in range(line):
            process_str = process_str + '|'
        for i in range(line, 101):
            process_str = process_str + ' '

        process_str = process_str + ']{}/{}'.format(index,train_full_len)
        sys.stdout.write(process_str)



f.close()
```
