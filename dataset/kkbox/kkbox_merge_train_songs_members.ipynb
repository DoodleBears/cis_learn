```python
songs_df = songs.drop(columns='Unnamed: 0',axis=1)
songs_df.head(3)
```

```python
train_df = train.drop(columns='Unnamed: 0',axis=1)
train_df.head(3)
```

```python
X_full = pd.merge(train_df, songs_df)
X_full
```
